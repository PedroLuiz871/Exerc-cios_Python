```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sns.countplot(x='APGAR1', data=sinasc)
plt.show()
```

```python
sns.countplot(x='APGAR5', data=sinasc)
plt.show()
```

```python
sns.countplot(x='SEXO', data=sinasc)
plt.show()
```

```python
sns.countplot(x='GESTACAO', data=sinasc)
plt.show()
```

```python
sns.countplot(x='ESTCIVMAE', data=sinasc)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0,0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1,1], x='CONSULTAS', data=sinasc)

plt.show()
```
