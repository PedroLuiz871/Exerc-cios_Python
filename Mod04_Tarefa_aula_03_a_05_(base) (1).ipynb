## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

df_mat['%_acertos'] = (df_mat['Qt_acertos'] / 24) * 100

df_por['%_acertos'] = (df_por['Qt_acertos'] / 18) * 100

df_geo['%_acertos'] = (df_geo['Qt_acertos'] / 8) * 100

df_ing['%_acertos'] = (df_ing['Qt_acertos'] / 8) * 100

df_his['%_acertos'] = (df_his['Qt_acertos'] / 8) * 100

df_fis['%_acertos'] = (df_fis['Qt_acertos'] / 12) * 100

df_qui['%_acertos'] = (df_qui['Qt_acertos'] / 12) * 100
````

````python
# 2) Seu código aqui
total_acertos = (

df_mat['Qt_acertos'] +

df_por['Qt_acertos'] +

df_geo['Qt_acertos'] +

df_ing['Qt_acertos'] +

df_his['Qt_acertos'] +

df_fis['Qt_acertos'] +

df_qui['Qt_acertos']

)

df_total = pd.DataFrame({'Total_acertos': total_acertos})
````

````python
# 3) Seu código aqui
total_pontos_possiveis = 24 + 18 + 8 + 8 + 8 + 12 + 12

df_total['%_geral'] = (df_total['Total_acertos'] / total_pontos_possiveis) * 100
````

````python
# 4) Seu código aqui
alunos_acima_corte = df_total[df_total['Total_acertos'] > 45].shape[0]

print(f"Total de alunos com nota acima de 45: {alunos_acima_corte}")

df_total.head()
````

````python
df_desqualificados = pd.DataFrame({

   'Matemática': df_mat['Qt_acertos'],

   'Física': df_fis['Qt_acertos'],

   'Química': df_qui['Qt_acertos']

})

df_desqualificados = df_desqualificados[

   (df_desqualificados['Matemática'] == 0) |

   (df_desqualificados['Física'] == 0) |

   (df_desqualificados['Química'] == 0)

]
df_desqualificados['Aluno_ID'] = df_desqualificados.index
df_desqualificados.head()
````

````python
num_desqualificados = df_desqualificados.shape[0]
print(f"Número de alunos desqualificados: {num_desqualificados}")
````

````python
indices_desqualificados = df_desqualificados.index
notas_hist_geo = pd.DataFrame({

   'História': df_his.loc[indices_desqualificados, 'Qt_acertos'],

   'Geografia': df_geo.loc[indices_desqualificados, 'Qt_acertos']

})
media_hist_geo = notas_hist_geo.mean()

print(f"Média em História: {media_hist_geo['História']:.2f}")

print(f"Média em Geografia: {media_hist_geo['Geografia']:.2f}")
````

````python
alunos_qualificados = ~df_total.index.isin(df_desqualificados.index)
df_segunda_fase = df_total[alunos_qualificados & (df_total['Total_acertos'] > 45)]
print(f"Número de alunos que passaram para a segunda fase: {df_segunda_fase.shape[0]}")

df_segunda_fase.head()
````
