1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
previsao = pd.read_csv("previsao_de_renda.csv")
previsao.head()
````

````python
previsao['data_ref'] = pd.to_datetime(previsao['data_ref'])
````

````python
ax = sns.countplot(x='data_ref',  hue = 'posse_de_veiculo', data=previsao)

tick_labs = previsao['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(previsao['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(-0.30, 1), loc=1, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'posse_de_imovel', data=previsao)

tick_labs = previsao['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(previsao['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(-0.30, 1), loc=1, borderaxespad=0.)
````

````python
#A diferença dos gráficos é grande e parece conter informaçao errada
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_veiculo', data=previsao, dodge=True, ci = 95)

tick_labs = previsao['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(previsao['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(-0.30, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_imovel', data=previsao, dodge=True, ci = 95)

tick_labs = previsao['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(previsao['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(-0.30, 1), loc=2, borderaxespad=0.)
````
